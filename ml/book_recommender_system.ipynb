## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name, cached_recommendations=None):
    if cached_recommendations is not None and book_name in cached_recommendations:
        return cached_recommendations[book_name]
    else:
        index = np.where(pt.index==book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

        data = []
        for i in similar_items:
            item = []
            temp_df = books[books['Book-Title'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

            data.append(item)

        if cached_recommendations is None:
            cached_recommendations = {}
        cached_recommendations[book_name] = data
        return data
```

## Now my approach

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity_scores = cosine_similarity(pt.T)
```

```python
def recommend_books_for_user(user_id):
    user_index = pt.columns.get_loc(user_id)

    similar_users = sorted(list(enumerate(user_similarity_scores[user_index])), key=lambda x: x[1], reverse=True)[1:5]

    recommended_books = []

    for similar_user_index, similarity_score in similar_users:
        user_ratings = pt.iloc[:, user_index]
        similar_user_ratings = pt.iloc[:, similar_user_index]

        unrated_books = similar_user_ratings[(user_ratings == 0) & (similar_user_ratings > 0)]

        recommended_books.extend(unrated_books.index)

    return recommended_books
```

```python
user_id = 254
recommended_books = recommend_books_for_user(user_id)
print(recommended_books)
```

## Reinforcement Learning

```python
import random

user_book_interactions = [
    (254, 17, 5),  # User 0 rated Book 1 with a rating of 5
    (254, 22, 4),
    (254, 24, 5),
    (254, 28, 1),
    (254, 25, 1),
]

num_episodes = 1000

num_users = len(users)
num_books = len(books)

Q_values = np.random.rand(num_users, num_books)

initial_epsilon = 0.5
final_epsilon = 0.01
epsilon_decay = 0.99

learning_rate = 0.1

def simulate_user_interaction():
    user_id, book_id, rating = random.choice(user_book_interactions)
    return user_id, book_id, rating

def select_action(state):
    epsilon = max(final_epsilon, initial_epsilon * (epsilon_decay ** episode))
    if np.random.rand() < epsilon:
        return np.random.randint(num_books)
    else:
        return np.argmax(Q_values[state])

for episode in range(num_episodes):
    user_id, book_id, rating = simulate_user_interaction()
    Q_values[user_id, book_id] += learning_rate * (rating - Q_values[user_id, book_id])
```

```python
user_id = 254
recommended_books = recommend_books_for_user(user_id)
print(recommended_books)
```
