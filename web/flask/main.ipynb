```python
import sqlite3
from time import time

def create_table(cursor):
    # cursor.execute('''
    #     DROP TABLE studentRating
    # ''')
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS studentRating (
            studentId VARCHAR(30),
            rating INT,
            time INT,
            PRIMARY KEY(studentId,rating,time)
        )
    ''')
    print("Table created successfully")


def insertRating(cursor, studentId, rating):


    try:


        cursor.execute(


            "INSERT INTO studentRating (studentId, rating,time) VALUES (?, ?, ?)", (studentId, rating,time()))
        return True


    except Exception as e:
        return e



def selectRating(cursor, studentId):


    try:


        cursor.execute(


            f"SELECT rating FROM studentRating WHERE studentId={studentId} ORDER BY time")


        rows = cursor.fetchall()
        data = []
        for row in rows:
            data.append(row[0])
        return data


    except Exception as e:
        return e



connection = sqlite3.connect('studentRating.db')


cursor = connection.cursor()
# create_table(cursor)
```

```python
insertRating(cursor,"studentId",1500)
```

```python
selectRating(cursor,"studentId")
```
